## Resolver:
## $$\mathrm y = ln(x-2)$$

```python
import numpy as np
import math
```

## 1. Método de la Bisección

```python
def f(x):
    return np.log(x-2)
```

```python
a = 2.5
b = 4
ni = 0
es = 0.01
ea = 1
rold = a
```

```python
while (es<ea):
      r = (a+b)/2
      ni = ni+1
      if (f(a)*f(r)<0):
         b=r
      else:
        if (f(r)*f(b)<0):
           a=r
      ea = abs((r-rold)/r)*100
      rold=r
print('El valor de la raiz por el Método de la Bisección es ',r)
```

## 2. Método de la Falsa Posición

```python
def f(x):
    return math.e**(-x)
```

```python
a = 2.5
b = 4
ni = 0
es = 0.01
ea = 1
rold = a
```

```python
while (es<ea):
      r = b - ((f(b)*(b - a))/(f(b) - f(a)))
      ni = ni+1
      if (f(a)*f(r)<0):
         b=r
      else:
        if (f(r)*f(b)<0):
           a=r
      ea = abs((r-rold)/r)*100
      rold=r
print('El valor de la raiz por el Método de la Falsa Posición es ',r)
```

## 3. Método de Newton-Rapshon

```python
def f(x):
    e = math.e
    return -e**(-x)
```

```python
def df(x):
    e = math.e
    return e**(-x)
```

```python
x1 = 4
ni = 6
es = 0.0001
ea = 1
```

```python
while (es<ea):
      x2 = x1 - f(x1)/df(x1)
      ni = ni+1
      ea = abs((x2-x1)/x2)*100
      x1=x2
print('El valor de la raiz por el Método de Newton-Rapshon es ',x2)
```

## 4. Método de la Secante

Define function `f`:

```python
def f(x):
    return np.log(x-2)
```

```python
a = 2.5
b = 4
ni = 0
es = 0.01
ea = 1
rold = a
```

```python
while (es<ea):
      r = b - (b-a)*f(b)/(f(b)-f(a))
      ni = ni+1
      ea = abs((r-rold)/r)*100
      a = b
      b = r
print('El valor de la raiz por el Método de la Secante es ',r)
```
